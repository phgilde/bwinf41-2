```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
for name in ["result99.txt", "result999.txt", "result9999.txt", "result99999.txt", "result999999.txt", "result9999999.txt"]:
    with open(name) as f:
        lines = f.readlines()[2:-2]
        x = []
        y = []
        for line in lines:
            line = line.split(", ")
            x.append(float(line[0]))
            y.append(float(line[1]))
        plt.plot(x, y, label=name)
        plt.yscale("log")
        plt.xscale("log")

plt.legend()
```
